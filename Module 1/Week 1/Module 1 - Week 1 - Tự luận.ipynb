```python
def exercise1(tp, fp, fn):
  name_list = ['tp', 'fp', 'fn']
  check_list =[isinstance(x, int) for x in [tp, fp, fn]]
  if all(check_list):
    if all( x > 0 for x in [tp, fp, fn]):
      precision = tp / (tp + fp)
      recall = tp / (tp + fn)
      f1 = 2*(precision * recall)/(precision + recall)

      print('Precision is {}'.format(precision))
      print('Recall is {}'.format(recall))
      print('F1 is {}'.format(f1))
    else:
      print('tp and fp and fn must be greater than zero')
  else:
    error_list= [name_list[i] for i in range(len(check_list)) if not check_list[i]]
    if len(error_list) == 1:
      print('{} must be int'.format(error_list[0]))
    else:
      temp_name = ''
      for i in range(len(error_list)):
        if i == 0:
          temp_name = temp_name + error_list[0]
        else:
          temp_name = temp_name + ' and ' + error_list[i]

      print('{} must be int'.format(temp_name))
```

Test

```python
exercise1(2,3,4)
exercise1('a',3,4)
exercise1(2,'a',4)
exercise1(2,3,0)
exercise1(2.1,2.2,5)
```

```python
import sys
import math
```

```python
def is_number(n):
  try:
    float(n)
    return True
  except ValueError:
    return False
```

```python
def sigmoid(x):
  return 1 / (1 + math.exp(-x))

def relu(x):
  if x > 0:
    return x
  else:
    return 0

def elu(x):
  if x > 0:
    return x
  else:
    return 0.01*(math.exp(x) - 1)
```

```python
def exercise2():
  x = input('Nhập giá trị x: ')
  if not is_number(x):
    print('x must be a number')
    sys.exit()

  x = float(x)

  function_list = ['sigmoid', 'relu', 'elu']
  act_func = input('Nhập tên activation function: ')
  if act_func not in function_list:
    print('{} is not supported'.format(act_func))
  elif act_func == 'sigmoid':
    print('sigmoid({}) = {}'.format(x, sigmoid(x)))
  elif act_func == 'relu':
    print('relu({}) = {}'.format(x, relu(x)))
  elif act_func == 'elu':
    print('elu({}) = {}'.format(x, elu(x)))
```

# BÀI 3

```python
import sys
import math
import numpy as np
import random
```

```python
def is_numeric(x):
  try:
    int(x)
    return True
  except ValueError:
    return False
```

```python
def MAE(predict, target):
  return np.abs(predict - target)

def MSE(predict, target):
  return (predict - target)**2

def RMSE(predict, target):
  return np.sqrt(MSE(predict, target))
```

```python
def exercise3():
  n = input('Nhập số lượng mẫu: ')
  if not is_numeric(n):
    print('num_samples must be an integer')
    sys.exit()

  target_list = [random.uniform(0, 1) for i in range(int(n))]
  predict_list = [random.uniform(0, 1) for i in range(int(n))]

  loss_list = ['MAE', 'MSE', 'RMSE']
  loss_name = input('Nhập tên loss function: ')
  if loss_name not in loss_list:
    print('{} is not supported'.format(loss_name))

  elif loss_name == 'MAE':
    loss_value = 0

    for i in range(int(n)):
      loss_value += MAE(predict_list[i], target_list[i])
      print('Loss name: MAE, sample: {}, pred: {}, target: {}, loss: {}'.format(i+1, predict_list[i], target_list[i], MAE(predict_list[i], target_list[i])))

    print('Final MAE: {}'.format(loss_value))
  elif loss_name == 'MSE':
    loss_value = 0

    for i in range(int(n)):
      loss_value += MAE(predict_list[i], target_list[i])
      print('Loss name: MSE, sample: {}, pred: {}, target: {}, loss: {}'.format(i+1, predict_list[i], target_list[i], MSE(predict_list[i], target_list[i])))

    print('Final MSE: {}'.format(loss_value))
  else:
    loss_value = 0

    for i in range(int(n)):
      loss_value += MAE(predict_list[i], target_list[i])
      print('Loss name: RMSE, sample: {}, pred: {}, target: {}, loss: {}'.format(i+1, predict_list[i], target_list[i], RMSE(predict_list[i], target_list[i])))

    print('Final RMSE: {}'.format(loss_value))
```

```python
exercise3()
```

```python
def factorial(n):
  temp = 1
  for i in range(1, n+1):
    temp *= i
  return temp
```

```python
def approx_sin(x, n):
  result = 0
  for i in range(n):
    result += ((-1)**i) * (x**(2*i+1)) / factorial(2*i+1)
  print(result)

def approx_cos(x, n):
  result = 0
  for i in range(n):
    result += ((-1)**i) * (x**(2*i)) / factorial(2*i)
  print(result)

def approx_sinh(x, n):
  result = 0
  for i in range(n):
    result += x**(2*i+1) / factorial(2*i+1)
  print(result)

def approx_cosh(x, n):
  result = 0
  for i in range(n):
    result += x**(2*i) / factorial(2*i)
  print(result)
```

```python
approx_sin ( x =3.14 , n =10)
approx_cos ( x =3.14 , n =10)
approx_sinh ( x =3.14 , n =10)
approx_cosh ( x =3.14 , n =10)
```

```python
def md_nre_single_sample(y,y_hat,n,p):
  print((y**(1/n) - y_hat**(1/n))**p)
```
