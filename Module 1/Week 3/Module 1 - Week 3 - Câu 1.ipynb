```python
import torch
import torch.nn as nn
```

Softmax

```python
class CustomSoftmax(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        x_exp = torch.exp(x)
        partition = x_exp.sum(0, keepdims=True)
        return x_exp / partition
```

```python
data = torch.Tensor ([1 , 2 , 3])
softmax = CustomSoftmax()
output = softmax(data)
output
```

Stable Softmax

```python
class CustomStableSoftmax(nn.Module):
    def __init__(self):
        super().__init__()

    def forward(self, x):
        x_max = torch.max(x, dim=0, keepdim=True)
        x_exp = torch.exp(x - x_max.values)
        partition = x_exp.sum(0, keepdims=True)
        return x_exp / partition
```

```python
data = torch . Tensor ([1 , 2 , 3])
softmax_stable = CustomStableSoftmax()
output = softmax_stable(data)
output
```
