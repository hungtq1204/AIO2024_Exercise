```python
def word_count(file_path):
    dict_count = {}

    with open(file_path, 'r') as file:
        words = file.read().split()

    for word in words:
        if word in dict_count.keys():
            dict_count[word] += 1
        else:
            dict_count[word] = 1

    return dict_count
```
