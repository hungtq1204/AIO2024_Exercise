Viết function trả về một dictionary đếm số lượng chữ xuất hiện trong một từ, với key là chữ cái là value là số lần xuất hiện.

1.   Input: một từ
2.   Output: dictionary đếm số lần các chữ xuất hiện
3.   Note: Giả sử các từ nhập vào đều có các chữ cái thuộc [a-z] hoặc [A-Z]





```python
def count_char(string):
  count_dict ={}

  for char in string:
    if char in count_dict.keys():
      count_dict[char] += 1
    else:
      count_dict[char] = 1

  count_dict = dict(sorted(count_dict.items(), key=lambda item: item[1]))
  return count_dict
```

```python
string = 'Happiness'
print(count_char(string))
```

```python
string = 'smiles'
print(count_char(string))
```
