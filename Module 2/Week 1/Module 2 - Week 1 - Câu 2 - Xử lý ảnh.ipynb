```python
import matplotlib.image as mpimg
import numpy as np
```

```python
def lightness_grayscale(vector):
  return np.max(vector)*0.5 + np.min(vector)*0.5

def average_grayscale(vector):
  return np.mean(vector)

def luminosity_grayscale(vector):
  return vector[0]*0.21 + vector[1]*0.72 + vector[2]*0.07
```

# Q12

```python
img = mpimg.imread('/content/dog.jpeg')
gray_img_01 = np.apply_along_axis(lightness_grayscale, axis=2, arr=img)
gray_img_01[0,0]
```

# Q13

```python
gray_img_02 = np.apply_along_axis(average_grayscale, axis=2, arr=img)
gray_img_02[0,0]
```

```python
gray_img_03 = np.apply_along_axis(luminosity_grayscale, axis=2, arr=img)
gray_img_03[0,0]
```
